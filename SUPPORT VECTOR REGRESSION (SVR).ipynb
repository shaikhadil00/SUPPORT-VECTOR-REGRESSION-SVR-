```python
x = df.iloc[:,0:1].values
x
```

```python
model.score(x,y)
```

```python
df = df.drop('age',axis=1)
```

```python
df.head()
```

```python
x = df.iloc[:,0:1].values
x
```

```python
y = df.iloc[:,1].values
y
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import StandardScaler
x_sc = StandardScaler()
x = x_sc.fit_transform(x)
y_sc = StandardScaler()
y = y_sc.fit_transform(np.reshape(y,(20,1)))
```

```python
from sklearn.svm import SVR
model = SVR(kernel = 'rbf')
model.fit(x,y)
```

```python
kernel :- convert low defination data into high defination data
rbf :- radio basic function
```

```python
plt.scatter(x,y, color ='b' , marker = 'D')
plt.plot(x,model.predict (x), color = 'r')
plt.xlabel('experience' , fontsize = 15)
plt.ylabel('income', fontsize = 15)
plt.title('SVR Graph', fontsize = 15)
plt.show
```

```python
model.score(x,y)
```
